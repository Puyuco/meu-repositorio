# Passos iniciais: Importando as Bibliotecas

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

# Leitura da Base de Dados

```python
base_dados =pd.read_csv( 'Salary_Data.csv' )
base_dados.head()
```

# Renomeando as colunas

```python
base_dados.rename(
columns = { 'YearsExperience' : 'Xp',
    'Salary' : 'Renda', 
} , inplace = True)
```

```python
base_dados.head()
```

# Verificando Dados Nulos

```python
sns.heatmap(base_dados.isnull());
```

```python
# nenhum dados nulos

base_dados.isnull().sum()
```

```python
base_dados.describe()
```

```python
plt.figure(figsize = (10,6))

plt.title('Anlise da Renda')

sns.kdeplot(base_dados['Renda'], color = 'green');
```

```python
plt.figure(figsize = (10,6))

plt.title('Anlise de Xp')

sns.kdeplot(base_dados['Xp'], color = 'blue');
```

```python
# Distribuio
plt.figure(figsize = (10,5))
sns.boxplot(base_dados['Renda']);
```

```python
# Distribuio
plt.figure(figsize = (10,5))
sns.boxplot(base_dados['Xp']);
```

```python
# Distribuio
plt.figure(figsize = (10,5))
sns.scatterplot(data = base_dados, x = 'Renda' , y = 'Xp');
```

```python
sns.regplot(data = base_dados, x = 'Renda' , y = 'Xp');
```

```python
# Correlao
correlacao = base_dados.corr('pearson')
plt.figure(figsize = (10,5))
sns.heatmap(correlacao, annot = True, cbar = False);
```

```python
# Correlao
correlacao = base_dados.corr('spearman')
plt.figure(figsize = (10,5))
sns.heatmap(correlacao, annot = True, cbar = False);
```
